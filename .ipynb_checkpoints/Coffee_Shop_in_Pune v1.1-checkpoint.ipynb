## Starbucks in Pune dataset by Webscrapping

```python
try:
    def eazydiner_webscrapping(pages_num):
        '''the function returns the dataframe containing address, ratings and coordinates for Starbucks stores in Pune'''
        rating_to_df= []
        address_to_df=[]
        latitude_df = []
        longitude_df = []
        for page in range(1, pages_num+1):
            response = requests.get("https://www.eazydiner.com/pune/restaurants/starbucks-coffee-pune"+"?total=20&page="+"{}".format(page)).text
            soup = BeautifulSoup(response, "html.parser")

            # Extracting rating for each store
            rating = soup.find_all(class_="critic")
            rate = []
            for j in range(len(rating)):
                rate.append(rating[j].text.strip())
            rating_to_df = rating_to_df+rate

            # Extracting locations names
            address=soup.find_all("h3", class_="margin-t-5 res_loc")
            addr = []
            for j in range(len(address)):
                addr.append(address[j].text.strip())
            address_to_df= address_to_df+addr

            # Extracting all directed links on the web page
            all_links= [var['href'] for var in soup.find_all('a', href=True)]
            map_links =[] 
            for num in range(338):
                if "/pune/starbucks" in all_links[num].strip():
                    if "map" in all_links[num].strip():
                        map_links.append(all_links[num].strip()[:-4])

            latitude = []
            longitude = []
            for cor in map_links:
                sub_response = requests.get("https://www.eazydiner.com"+cor).text
                soup_map = BeautifulSoup(sub_response, "html.parser")

                f=soup_map.find_all("script",{"type":"application/ld+json" })
                json_string = f[0].text.strip()
                obj= json.loads(json_string)
                lat = obj["@graph"][0]["geo"]["latitude"]
                latitude.append(lat)
                lon = obj["@graph"][0]["geo"]["longitude"]
                longitude.append(lon)

            latitude_df = latitude_df+ latitude
            longitude_df = longitude_df +longitude

        df=pd.DataFrame()
        df["Address"] = address_to_df
        df["Rating"]= rating_to_df
        df["Latitude"] = latitude_df
        df["Longitude"] = longitude_df

        return df

except:
    print('Please check your Internet connection')
```

```python
try:
    start = time.time()
    starbucks = eazydiner_webscrapping(2)
    end = time.time()
    elapsed = end-start

    print("time taken by the webscrapping function is: ", elapsed, " seconds")
    
    # Saving the file to the directory
    starbucks.to_csv("to_be_used/starbucks.csv")
    
except:
    print('Please check your internet connection but you can still enjoy the further project. This data is made available in offline mode alternatively.')
```

## Demographics

Demographic data is very crucial for our analysis. It can help to understand the neighborhood and draw some inferences in terms of coffee-drinking habits. What we are mainly looking for in our analysis, is a dataset containing the following information about Pune:

        1) Population of each zone
        2) Wards (Prabhag) data
        3) Area of each zone
        4) Geographic locations of each zone
        5) Prabhag wise demography (self modifies and processed dataset in another notebook)
        
The datasets are selected from different web-sources. It is observed that zone names are not uniformly spelled. Before we load different demographic datasets, lets create a function which can make uniform zone names. 

```python
def uniform_zone_names(x):
    """The function return the standard name for zones replacing the given name in the dataset"""
    
    if "Aundh" in x:
        return "Aundh"
    elif "Bhavani" in x or "Bhawani" in x:
        return "Bhawani_Path"
    elif "Bibve" in x or "Bibwe" in x:
        return "Bibwewadi"
    elif "Dhank" in x:
        return "Dhankawadi"
    elif "Dhole" in x:
        return "Dhole_Patil_Road"
    elif "Ghole" in x:
        return "Ghole_Road"
    elif "Hadap" in x or "Hadpsar" in x:
        return "Hadapsar"
    elif "Kasba" in x or "Kasaba" in x:
        return "Kasba_Vishrambagwada"
    elif "Koth" in x:
        return "Kothrud"
    elif "Sahakar" in x or "Sahkar" in x:
        return "Sahakarnagar"
    elif "Sangam" in x:
        return "Yerwada_Sangamwadi"
    elif "Tilak" in x or "Tillak" in x:
        return "Tilak_Road"
    elif "Warje" in x or "Waraje" in x:
        return "Warje_Karvenagar"
    elif "Nagar" in x:
        return "Nagar_Road"
    elif "Kondh" in x:
        return "Kondhwa_Wanawadi"
```

### Where are the Starbucks located?

```python
def map_locations_to_wards(location):
    '''the function maps coordinate and assign the ward(prabhag) name where the point coordinates lies '''
    
    for i in prabhag.index: 
        if prabhag.loc[i, "geometry"].contains(location):
            return prabhag.loc[i, "Prabhag_Name"]
```

```python
def map_locations_to_zones(location):
    '''the function maps coordinate and assign the zone name where the point coordinates lies '''
    
    for i in demography.index: 
        if demography.loc[i, "geometry"].contains(location):
            return demography.loc[i, "Zone_Name"]
```

```python
def plot_counts(df):
    '''the funtion retuns the counplots for the wards and zones in the dataset'''
    
    fig, ax=plt.subplots(1,2,figsize=(14,8))
    
    sns.countplot(df.Ward,ax=ax[0])
    ax[0].set_title("Wards".title(), fontsize= 12)
    ax[0].set_xticklabels(labels= df.Ward.unique(), rotation=90)
    
    sns.countplot(df.Zone, ax= ax[1])
    ax[1].set_title("Zones".title(), fontsize= 12)
    ax[1].set_xticklabels(labels= df.Zone.unique(), rotation=90)
```

```python
starbuck_wards = starbucks.Ward.unique()
decision_metrics= prabhag[prabhag['Prabhag_Name'].isin(starbuck_wards)][['population', 'Population_density']].mean().reset_index()
decision_metrics.columns= ['Attribute', 'with_Starbucks']

with_starbucks_ward =prabhag[prabhag['Prabhag_Name'].isin(starbuck_wards)]
no_starbucks_ward = prabhag[~prabhag['Prabhag_Name'].isin(starbuck_wards)]
decision_metrics['No_Starbucks']=no_starbucks_ward['population'].mean()

with_cafe= cafes[cafes['Ward'].isin(starbuck_wards)].count()['name']/with_starbucks_ward['area'].sum()
no_cafe= cafes[~cafes['Ward'].isin(starbuck_wards)].count()['name']/no_starbucks_ward['area'].sum()

cafe_den=dict(Attribute='Cafe_density', 
              with_Starbucks = with_cafe,
              No_Starbucks=no_cafe)

decision_metrics= decision_metrics.append(cafe_den, ignore_index=True)

with_attract= attractions[attractions['Ward'].isin(starbuck_wards)].count()['Name']/with_starbucks_ward['area'].sum()
no_attract= attractions[~attractions['Ward'].isin(starbuck_wards)].count()['Name']/no_starbucks_ward['area'].sum()

attract_den=dict(Attribute='Attraction_density', 
              with_Starbucks=with_attract,
                No_Starbucks=no_attract)

decision_metrics= decision_metrics.append(attract_den, ignore_index=True)

with_college= colleges[colleges['Ward'].isin(starbuck_wards)].count()['name']/with_starbucks_ward['area'].sum()
no_college= colleges[~colleges['Ward'].isin(starbuck_wards)].count()['name']/no_starbucks_ward['area'].sum()

college_den=dict(Attribute='College_density', 
                 with_Starbucks=with_college,
                No_Starbucks= no_college)

decision_metrics= decision_metrics.append(college_den, ignore_index=True)

decision_metrics
decision_metrics.style.background_gradient(subset=['with_Starbucks', 'No_Starbucks'], 
                                  cmap='flare', axis=1)
```

```python
three_starbucks= prabhag[prabhag['Prabhag_Name'].isin(three_outlets)][['population', 'Population_density']].mean().reset_index()
three_starbucks.columns= ['Attribute', 'Three_Outlets']
two_starbucks= prabhag[prabhag['Prabhag_Name'].isin(two_outlets)][['population', 'Population_density']].mean().reset_index()
two_starbucks.columns= ['Attribute', 'Two_Outlets']
one_starbucks= prabhag[prabhag['Prabhag_Name'].isin(one_outlet)][['population', 'Population_density']].mean().reset_index()
one_starbucks.columns= ['Attribute', 'One_Outlet']

decision_metrics = one_starbucks.merge(two_starbucks, on='Attribute')
decision_metrics = decision_metrics.merge(three_starbucks, on='Attribute')

three_cafe_den=cafes[cafes['Ward'].isin(three_outlets)].count()['name']/(8.21+14.96)
two_cafe_den=cafes[cafes['Ward'].isin(two_outlets)].count()['name']/(10.15+4.15+4.30)
one_cafe_den=cafes[cafes['Ward'].isin(one_outlet)].count()['name']/(2.19+4.15+1.46)

cafe_den=dict(Attribute='Cafe_density', 
              One_Outlet= one_cafe_den, 
              Two_Outlets=two_cafe_den, 
              Three_Outlets=three_cafe_den)

decision_metrics= decision_metrics.append(cafe_den, ignore_index=True)

three_attract_den=attractions[attractions['Ward'].isin(three_outlets)].count()['Name']/(8.21+14.96)
two_attract_den=attractions[attractions['Ward'].isin(two_outlets)].count()['Name']/(10.15+4.15+4.30)
one_attract_den=attractions[attractions['Ward'].isin(one_outlet)].count()['Name']/(2.19+4.15+1.46)

attract_den=dict(Attribute='Attraction_density', 
              One_Outlet= one_attract_den, 
              Two_Outlets=two_attract_den, 
              Three_Outlets=three_attract_den)

decision_metrics= decision_metrics.append(attract_den, ignore_index=True)

three_college_den=colleges[colleges['Ward'].isin(three_outlets)].count()['name']/(8.21+14.96)
two_college_den=colleges[colleges['Ward'].isin(two_outlets)].count()['name']/(10.15+4.15+4.30)
one_college_den=colleges[colleges['Ward'].isin(one_outlet)].count()['name']/(2.19+4.15+1.46)

college_den=dict(Attribute='College_density', 
              One_Outlet= one_college_den, 
              Two_Outlets=two_college_den, 
              Three_Outlets=three_college_den)

decision_metrics= decision_metrics.append(college_den, ignore_index=True)

decision_metrics
decision_metrics.style.background_gradient(subset=['One_Outlet', 'Two_Outlets', 'Three_Outlets'], 
                                  cmap='flare', axis=1)
```

# <font color=darkred> **Site Selection for our Coffee Shop**

With our long journey so far, we have now come to the part where we can decide on a selection neighborhood for our coffee shop based on the analysis done in the previous sections. The factors such as demography, localities with tourism spots, universities, and competitors are the decisive elements for site selection. We have also tried to dissect the strategy of Starbucks in selecting locations in Pune. Though Starbucks overlooked the local competitors, we will select the optimum restaurant density spots for our selection. Other factors that mainly drive our decision are as follows:

        Population desity
        Cafe density
        College density 
        Neighborhood with tourism spots

Our next step is to find the above values for each of the zones. The competition and population should be low in our selected area, but the other two characteristics are aimed to be high. Since the above factors serve their roles in our decision-making, we will go with the weighted average method to select a neighborhood with the optimum value.

Since we are finding the density i.e. dividing the number of entities by area. For some wards such as the City post, the area is very small (0.45 Km2), hence it is causing an explosion in density parameters. To avoid this, we will select the ward for which the area is above its median value i.e. 2.175 Km2. Cafe density values typically represent how closed the restaurant is in the region. We want to have fewer competitors around us. So we will consider the factor (1-cafe density), also called cafe disperseness. As understood from the requirement, the cafe needs to be located where the colleges and attractions are more and population, population density, and competition is moderate. Hence weight values are assigned as follows:


           #Attributes              #Weights
           
        Population                     1
        
        Population density             1
        
        Cafe Disperseness              1
        (1- Cafe density)
        
        College density                2
        
        Attractions density            2
        
     
        
        

```python
prabhag_selected = prabhag[prabhag["area"]>= prabhag["area"].median()]
prabhag_selected.shape
```

```python
def attribute_ward_density(attribute_df):
    '''the function returns the density of attributes i.e number of entities per square kilometer'''
    
    attribute_density_dict={}
    for ward in attribute_df['Ward'].unique():
        
        if ward in list(prabhag_selected["Prabhag_Name"]):

            num= (attribute_df['Ward'].value_counts()[ward])
            den= prabhag_selected.loc[prabhag_selected[prabhag_selected['Prabhag_Name']==ward].index[0], 'area']
            attribute_ward_density= num/den
            attribute_density_dict[ward]= round(attribute_ward_density,4)
    return attribute_density_dict
```

```python
cafeward= attribute_ward_density(cafes)
df = pd.DataFrame({'Prabhag_Name': list(cafeward.keys()), 'Cafe_density': list(cafeward.values())})
attractward=attribute_ward_density(attractions)
df1 = pd.DataFrame({'Prabhag_Name': list(attractward.keys()), 'Attraction_density': list(attractward.values())})
collegeward= attribute_ward_density(colleges)
df2 = pd.DataFrame({'Prabhag_Name': list(collegeward.keys()), 'College_density': list(collegeward.values())})

density= prabhag[['Prabhag_Name','population', 'Population_density']]

overall_ward= density.merge(df, on='Prabhag_Name', how="left")
overall_ward= overall_ward.merge(df1, on='Prabhag_Name', how="left", )
overall_ward= overall_ward.merge(df2, on='Prabhag_Name', how= "left")
overall_ward.fillna(0, inplace=True)

scaler = MinMaxScaler()
overall_ward[['population', 'Population_density', 'Cafe_density', 'Attraction_density', 'College_density']] = scaler.fit_transform(overall_ward[['population', 'Population_density', 'Cafe_density', 'Attraction_density', 'College_density']])

overall_ward['weighted_average'] = (1*(1-overall_ward['Cafe_density'])+2*overall_ward['Attraction_density']+2*overall_ward['College_density']+1*overall_ward['population']+1*overall_ward['Population_density'])/5
overall_ward.sort_values(by='weighted_average', ascending=False, ignore_index=True, inplace=True)
overall_ward['rank'] = overall_ward['weighted_average'].rank(ascending=False).astype('int')
prabhag_geo= prabhag[['Prabhag_Name', 'geometry']]
overall_ward= overall_ward.merge(prabhag_geo, on='Prabhag_Name', how='left')
overall_ward=gpd.GeoDataFrame(overall_ward, geometry='geometry')

top_wards = overall_ward.iloc[:3,:]
```
